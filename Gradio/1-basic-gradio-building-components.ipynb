  This notebook primarily showcases the few building blocks of the <b>Gradio</b> web applications building tool. ### <span style="color:#123524;  font-weight: bold;">The Interface class</span>

<p style="text-align: justify; text-justify: inter-word; font-size:17px;">
  The gr.Interface class is a high-level abstraction in Gradio that allows you to quickly create a demo for any Python function simply by specifying the input types and the output types.<br> We see that the Interface class is initialized with three required parameters:
    <ul style="text-align: justify; text-justify: inter-word; font-size:17px;">
        <li><b>fn</b>: the function to wrap a user interface (UI) around</li>
        <li><b>inputs</b>: which Gradio component(s) to use for the input. The number of components should match the number of arguments in your function.</li>
        <li><b>outputs</b>: which Gradio component(s) to use for the output. The number of components should match the number of return values from your function.</li>
    </ul>
</p>

### <span style="color:#123524;  font-weight: bold;">Gradio Components</span>

<p style="text-align: justify; text-justify: inter-word; font-size:17px;">
 Gradio includes more than 30 pre-built components (as well as many community-built custom components) that can be used as inputs or outputs in your demo. These components correspond to common data types in machine learning and data science, e.g. the gr.Image component is designed to handle input or output images, the gr.Label component displays classification labels and probabilities, the gr.LinePlot component displays line plots, and so on.
</p>

```python
import gradio as gr
```

```python
gr.__version__
```

```python
# Defining a simple sum function
def add_numbers(a, b):
    return a+b

interface = gr.Interface(fn=add_numbers,
                         inputs = [gr.Number(placeholder='Enter an integer value'),
                                   gr.Number(placeholder='Enter an integer value')],
                         outputs = gr.Number(label="a + b ="),
                         theme = "monochrome")
interface.launch()
```
